```python
import os
import cv2
import numpy as np
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
import joblib
import insightface

# 1. Inisialisasi InsightFace
model = insightface.app.FaceAnalysis(providers=['CUDAExecutionProvider', 'CPUExecutionProvider'])
model.prepare(ctx_id=0, det_size=(640, 640))
```

```python
img = cv2.imread("WhatsApp Image 2025-12-10 at 16.28.02_cb4337b5.jpg")
clf = joblib.load("face_recognition_model.pkl")


faces = model.get(img)
if len(faces) > 0:
        # Get embeddings for all detected faces
        embeddings = []
        for i, face in enumerate(faces):
            embedding = face.embedding
            embeddings.append(embedding)
            print(f"Face {i+1} embedding shape: {embedding.shape}")
        
        # Make predictions
        embeddings_array = np.array(embeddings)
        predictions = clf.predict(embeddings_array)
        
        # Get prediction probabilities
        probabilities = clf.predict_proba(embeddings_array)
        
        # Load label encoder to get class names
        le = LabelEncoder()
        
        print("\n=== Predictions ===")
        for i, (pred, probs) in enumerate(zip(predictions, probabilities)):
            print(f"\nFace {i+1}:")
            print(f"  Predicted Class ID: {pred}")
            print(f"  Confidence Score: {max(probs):.4f}")
```
